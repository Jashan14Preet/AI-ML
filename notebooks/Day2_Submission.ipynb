**Vectorization:** It is a technique in data processing , used in NumPy, in which operations are directly applied to the whole array or dataset at once, without traversing and processing each element using loops one by one.

*For Example,* in numpy_task.py we scaled by doubling all the values of the matrix without using a loop.
*i.e. processed_data = data_points * 2*  
Here, data_points is reshaped into a 2X2 matrix, and the operation is applied to all the elements at once.


NumPy is used here as the reshaped matrix allows the model to process the given data in parallel,thus makes the overall computations faster and more efficient
