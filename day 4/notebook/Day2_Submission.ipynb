# Day 4: Pre-processing Techniques

```python
import pandas as pd
import numpy as np

data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [21, np.nan, 23, 24],
    'Score': [80, 75, np.nan, 95]
}

df = pd.DataFrame(data)
df
```

```python
# Dropping a row
df_dropped = df.dropna()
df_dropped
```

```python
#Filing a row
df_filled = df.copy()

df_filled['Age'] = df_filled['Age'].fillna(df_filled['Age'].mean())
df_filled['Score'] = df_filled['Score'].fillna(0)

df_filled
```

### Comparison:

- Dropping data removes incomplete rows but reduces dataset size.
- Filling data keeps all records and maintains dataset size.
- Imputation is preferred when data is limited.

### 2AZWhy is it better to fill a missing age with the mean rather than just putting 0?
If we replace missing age with 0, it will be impossible since age cannot be zero.
This will introduce erroneous information to our analysis.

If we replace missing age with the mean value, it will give us a sensible replacement.
We can maintain the structure of our data.

Thus, using the mean for filling missing age is preferable over using 0.
